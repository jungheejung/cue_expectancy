```python
from nilearn import datasets, plotting, image
from nilearn.input_data import NiftiLabelsMasker
from nilearn.image import load_img, resample_to_img
import pandas as pd

# from cmcrameri import cm
import matplotlib.pyplot as plt
import numpy as np
```

# load atlas

```python
roi_list = ['npsneg_lLOC', 'npsneg_lSTS', 'npsneg_PCC', 'npsneg_pgACC', 
            'npsneg_rIPL', 'npsneg_rLOC', 'npsneg_rpLOC',
            'npspos_dACC', 'npspos_lIns', 'npspos_rdpIns', 'npspos_rIns', 'npspos_rS2_Op', 
            'npspos_rThal', 'npspos_rV1']
target_img = datasets.load_mni152_template(resolution=3)
for roi in roi_list:
    roi_fname = f'/Users/h/Documents/projects_local/cue_expectancy/data/atlas/{roi}.nii'
    roi_img = load_img(roi_fname)
    roi_resampled = image.resample_to_img(source_img=roi_img, target_img=target_img, interpolation='nearest')
    c = plotting.plot_stat_map(roi_resampled, cmap='Reds', draw_cross=False, title=roi)
    c.savefig(f'/Users/h/Documents/projects_local/cue_expectancy/resources/atlas/{roi}.png')
```
