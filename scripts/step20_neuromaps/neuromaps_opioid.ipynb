In this markdown, I grab the receptor maps to check whether the low-cue-high-NPS effect is related to threat analgesia and the endogenous opioid system

```python
import nibabel as nib
import numpy as np
import neuromaps
from neuromaps import datasets

from neuromaps.datasets import fetch_fslr

# from neuromaps.datasets import fetch_fsaverage
from surfplot import Plot
from nilearn.plotting.cm import _cmap_d as nilearn_cmaps
```

```python
abagen = datasets.fetch_annotation(source='abagen', desc='genepc1')
print(abagen)
```

```python
# fslr example
surfaces = fetch_fslr()
lh, rh = surfaces['inflated']

p = Plot(surf_lh=lh, surf_rh=rh)
fig = p.build()
fig.show()
```

## Kantonen 2020: PET tracer binding (BPnd) to MOR (mu-opioid receptor) opioid

```python
kantonen2020 = datasets.fetch_annotation(source=['kantonen2020', 'carfentanil', 'MNI152', '3mm'])

kantonen_lh, kantonen_rh = neuromaps.transforms.mni152_to_fslr(kantonen2020, fslr_density='32k', method='linear')

surfaces = fetch_fslr()
lh, rh = surfaces['inflated']
sulc_lh, sulc_rh = surfaces['sulc']
# color_range = (-12, 12)

# p = Plot(lh, rh)
p = Plot(surf_lh=lh, surf_rh=rh, size=(1000, 200), zoom=1.2, layout='row', views=['lateral', 'medial', 'ventral', 'posterior'], mirror_views=True)

p.add_layer({'left': sulc_lh, 'right': sulc_rh}, cmap='binary_r', cbar=False)

#   -> add statistical maps
p.add_layer({'left': kantonen_lh, 'right': kantonen_rh}, cmap='viridis',
            # color_range=color_range, 
            cbar_label='Kantonen_2020')
cbar_kws = dict(outer_labels_only=True, pad=.02, n_ticks=2, decimals=0)

#   -> build figure
fig = p.build(cbar_kws=cbar_kws)
fig.axes[0].set_xlabel('z', labelpad=-11, fontstyle='italic')
fig.show()
```
