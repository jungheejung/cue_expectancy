## 1-1. Stimulus Intensity contrast & covariates (raw)
Here, we look at the effect of including WM, CSF covariates in a multiple regression. <br>
Does the inclusion of WMCSF help improve power of the group average and behavioral covariate maps?
* Contrast map: 'P_simple_STIM_stimlin_high_gt_low'
* covariate: Behavioral stim effect (Raw)
* covariate: include WMCSF vs. not

```matlab
% Set the default Size for all new figures
set(groot, 'defaultFigurePosition', [100, 100, 800, 600]);

% set(groot, 'defaultFigureRenderer', 'painters');
main_dir = '/Users/h/Documents/projects_local/cue_expectancy';
save_dir = fullfile(main_dir, 'analysis', 'fmri', 'spm', 'scaling','6conditions_highlowcue_rampplateau'); 
contrast_of_interest = 'P_simple_STIM_stimlin_high_gt_low';
load(fullfile(save_dir, ['tobj_cov-rawstim_rating-outcome_contrast-', contrast_of_interest, '_csfwm-true.mat']));
load(fullfile(save_dir, ['tobj_cov-rawstim_rating-outcome_contrast-', contrast_of_interest, '_csfwm-false.mat']));

fontsize = 12;
figure;
x = stim_raw_parcel_nullcov.tscores(:,2);
y = stim_raw_parcel_csfwmcov.tscores(:,4); % intercept

scatter(x, y);

% Calculate min and max for axes
minVal = min(min(x), min(y));
maxVal = max(max(x), max(y));
axisRange = [minVal - 1, maxVal + 1]; % Extend the range slightly for visual clarity
%plot inline
% Set axes limits
xlim(axisRange);
ylim(axisRange);

axis equal;


% Drawing lines
line(xlim, [3 3], 'Color', 'k', 'LineStyle', '--', 'LineWidth', 1); % Horizontal line at y = 3
line(xlim, [-3 -3], 'Color', 'k', 'LineStyle', '--', 'LineWidth', 1); % Horizontal line at y = -3
line([3 3], ylim, 'Color', 'k', 'LineStyle', '--', 'LineWidth', 1); % Vertical line at x = 3
line([-3 -3], ylim, 'Color', 'k', 'LineStyle', '--', 'LineWidth', 1); % Vertical line at x = -3

set(gca, 'FontSize', fontsize);

% Adding labels and title
xlabel({'t-values of Stimulus intensity contrast', 'Intercept (raw) NO CSFWM'}, 'FontSize', fontsize);

ylabel({'t-values of Stimulus intensity contrast', 'Intercept (raw) including CSFWM covariate)'}, 'FontSize', fontsize);

title({'Does inclusion of WM/CSF lead to a difference',...
       'in Stimulus intensity group average maps?', ...
       '- covariate: behavioral stimulus intensity effects with raw rating', ...
       '- covariate: include wm/csf vs. not'}, 'FontSize', fontsize);

```

```matlab
montage(stim_raw_parcel_csfwmcov.t_obj)
```

```matlab
montage(stim_raw_parcel_nullcov.t_obj)
```
